```python
from IPython.core.display import display, update_display, HTML
import time, sys
```

<h3>Simple Emoji Animation</h3>

```python
# Clock Frames
clock = list(range(128336,128359))
# Moon Frames
moon = list(range(127761, 127768))
# Index for clock and moon frames
i = 0
j = 0
run = True
# Initialize Display and set an ID
display(HTML("<span style='font-size: 3rem;'>&#" + str(clock[i]) + ";</span>") ,display_id = "clock")
display(HTML("<span style='font-size: 3rem;'>&#" + str(moon[i]) + ";</span>") ,display_id = "moon")
while run:
    try:
        # Update the display with the next frame
        update_display(HTML("<span style='font-size: 3rem;'>&#" + str(clock[i]) + ";</span>"), display_id = "clock")
        update_display(HTML("<span style='font-size: 3rem;'>&#" + str(moon[j]) + ";</span>"), display_id = "moon")
        # Increment Frame Indexes
        i+=1
        j+=1
        # Reset Frame Indexes if they've exceeded the number of frames in the lists
        if(i >= len(clock)):
            i = 0
        if(j >= len(moon)):
            j = 0
        time.sleep(0.1)
        # Stop Running if the Kernel is interrupted.
    except KeyboardInterrupt:
        run = False
   
```
